```python
data2 = np.random.rand(20,3)
col = ['X','Y','Z']
df2 = pd.DataFrame(data = data2 , columns= col)
df2.head()
```

   ## Plot()
   You can plot different plots using the plot() function of pandas
   - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot
    - 'hexbin' : hexbin plot.

### Histogram

```python
df1['A'].hist()
```

```python
df1['A'].hist(bins = 50)   # Histogram now consists of 50 points
```

```python
# Histogram for all the columns in the data frame 
df1.plot(kind = 'hist' , bins = 50 , alpha = 0.7)
```

```python
# Stacked Histogram 
df1.plot(kind = 'hist' , bins = 50 ,  )
```

```python
# Horizontal Stacked Histogram 
df1.plot(kind = 'hist' , bins = 50 , alpha = 0.7 , stacked = True , orientation = 'horizontal')
```

### KDE Plot

```python
df1['A'].plot( kind = 'kde')
```

```python
# Kde Plot for all the colimns in the data frame
df1.plot(kind = 'kde')
```

### Area Plot

```python
#df1.plot(kind = 'area')
# Error : : When stacked is True, each column must be either all positive or negative.A contains both positive and negative values

# Alpha is used to make the plot transparent 
df2.plot(kind = 'area' , title = 'Area Plot', alpha = 0.5 , legend = True)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
df2.plot(kind = 'area' , title = 'Area Plot', legend = True)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

### Line Plot

```python
df1.plot(kind = 'line' )
```

### Scatter Plot

```python
# Scatter plot requires numerical columns for x and y
df1.plot(kind = 'scatter' , x = 'A' , y = 'B')
```

```python
# Multiplt plots on the same axis
ax = df1.plot(kind = 'scatter' , x = 'A' , y = 'B')
```

```python
ax = df1.plot(kind = 'scatter' , x = 'A' , y = 'B')
df1.plot(kind = 'scatter' , x = 'C' , y = 'D' , color = 'yellow' , ax = ax)
```

```python
# 'C' is the third column. Plot the scatter plot based on the third column in comparison with column A and column B
df1.plot(kind = 'scatter' , x = 'C' , y = 'D' , c = 'C')
```

```python
# Changing the color style using cmap = 'Rainbow'
df1.plot(kind = 'scatter' , x = 'C' , y = 'D' , c = 'C' , cmap = 'rainbow')
```

```python
# Changing the size of the symbols to 50
df1.plot(kind = 'scatter' , x = 'C' , y = 'D' , s = 50 ,c = 'C' , cmap = 'rainbow')
```

```python
# We can also specify the the size of the bubbles(dots) in the plot based on the a specific column value
# For eg if the df['C'][1] = 22 , df['C'][2] = 52, the size of the bubble or dot in the plot for  df['C'][2] = 52 will be 
# larger than the size of the bubble/dot for  df['C'][1] = 22
```

```python
df1.plot(kind = 'scatter' , x = 'A' , y = 'B' , s = df1['C']*100 , color = 'orange')
```

### Box Plot

```python
df2.plot(kind = 'box')
```

```python
# df2.plot(kind = 'box' , orientation = 'horizontal')
# You can't use orientation paramter to make the box plot horizontal . Instead use vert parameter
df2.plot(kind = 'box' , vert = False)
```

```python
# Changing the color of the box plot elements

color = dict(boxes = 'DarkGreen' , whiskers = 'DarkOrange' , medians = 'DarkBlue' , caps = 'Grey')
df2.plot(kind = 'box' , color = color)
```

### Hex Plot

```python
df1.plot(kind = 'hexbin' , x = 'A' , y = 'B')
```

```python
# Grid size is used to control the size of the hexagon in the output plot
df1.plot(kind = 'hexbin' , x = 'A' , y = 'B' , gridsize = 25 , color = 'orange')
```

```python
df1.plot(kind = 'hexbin' , x = 'A' , y = 'B' , gridsize = 25 , cmap = 'Oranges')
```

### Pie Plot

```python
# For Pie plot if there are NaN values it will be automatically be converted to 0 
# And for Pie plot there shouldn't be any negative values

df2.plot( y = 'X' ,kind = 'pie' , figsize = (8,8) , legend = False)
```

```python
# Instead of only column X , what if we want tp plot for all the columns in the data frame
df2.plot( subplots = True ,kind = 'pie' , figsize = (12,8) , legend = False)
```
