# Figure Asthetics

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

iris = sns.load_dataset('iris')
iris.head()
```

## Set Style
We can set particular styles using `set_style()`. 
There are five preset seaborn `themes: darkgrid, whitegrid, dark, white, and ticks`. They are each suited to different applications and personal preferences.  

```python
g = sns.swarmplot(x = 'sepal_length' , y = 'petal_length' , hue = 'species' ,data = iris ) 
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
# Tics to the axis
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
g = sns.swarmplot(x = 'sepal_length' , y = 'petal_length' , hue = 'species' ,data = iris , palette= 'coolwarm') 
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
# darkgrid
sns.set_style('darkgrid')
g = sns.swarmplot(x = 'sepal_length' , y = 'petal_length' , hue = 'species' ,data = iris , palette= 'coolwarm') 
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
#whitegrid
sns.set_style('whitegrid')
g = sns.swarmplot(x = 'sepal_length' , y = 'petal_length' , hue = 'species' ,data = iris , palette= 'coolwarm') 
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

## Removing axes spines - despine( )

```python
sns.countplot(x= 'species' , data = iris)
```

```python
sns.countplot(x= 'species' , data = iris)
# Removing top and right axis line
sns.despine(top=True, right=True, left=False, bottom=False,)
```

 ## Figure size and Aspect

```python
#We can simply use matplotlib's *plt.figure(figsize=(width,height) * to change the size of most seaborn plots.
#We can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect
sns.scatterplot(x= 'sepal_length' , y= 'petal_length' , hue= 'species' , data= iris)
```

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x= 'sepal_length' , y= 'petal_length' , hue= 'species' , data= iris)
```

## Set Context

```python
# Paper
plt.figure(figsize=(12,7))
sns.set_context('paper',font_scale=1)
sns.scatterplot(x= 'sepal_length' , y= 'petal_length' , hue= 'species', palette= 'rainbow',style= 'species' , size = 'species' , data= iris)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
sns.set_context('notebook')
sns.scatterplot(x= 'sepal_length' , y= 'petal_length' , hue= 'species', palette= 'rainbow',style= 'species' , data= iris)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
# poster 
plt.figure(figsize=(12,7))
sns.set_context('poster')
sns.scatterplot(x= 'sepal_length' , y= 'petal_length' , hue= 'species', palette= 'rainbow',style= 'species' , data= iris)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```
