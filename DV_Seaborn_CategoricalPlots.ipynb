# Seaborn (Categorical Plots) 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

## Stip Plot

```python
sns.stripplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
# No data for days  (Mon ,Tue , Wed)
tips[(tips['day'] == 'Mon') | (tips['day'] == 'Tue') | (tips['day'] == 'Wed')]
```

```python
# GItter is the position of data along the axis
# Default is true
sns.stripplot(x = 'day' , y = 'total_bill' , jitter = False , data = tips)
```

```python
# We can specify 3 variable in categorical data
sns.stripplot(x = 'day' , y = 'total_bill' , hue = 'sex' ,data = tips )
plt.legend(loc = 2)               # Will Place the legend in top left corner
```

```python
# Plotting the legend outside the scatter plot box
sns.stripplot(x = 'day' , y = 'total_bill' , hue = 'sex' ,data = tips )
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

## Swarm Plot

```python
sns.swarmplot(x = 'day' , y = 'total_bill' , hue = 'sex' ,data = tips)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
#The plotting of category should generally be along the X axis
sns.swarmplot(y = 'day' , x = 'total_bill' , hue = 'time' ,data = tips)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)

# Almost on thursday all the people come for lunch
# People refrain from eating outside on Friday
```

## Box Plot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. 

The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.


The box plot has 4 quartiles
1. Quartile1 (Betweem the lower whiser and the bottom of the box . The bottom of the box indicates the lowest value among the boxed data) 
2. Quartile2 (Between the bottom of the box and the median line{which is the mean of the boxed data})
3. Quartile3 (Between the median line and the top of the box . The top of the box indicates the highest value among the boxed data)
1. Quartile4 (Betweem the top whiser and the top of the box) 

```python
sns.boxplot(x = 'day' , y = 'total_bill' ,data = tips)
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , hue = 'sex', palette= 'rainbow' ,data = tips)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

## Violin Plot

The Violin Plot is a combination of Box plot and KDE surrounding the box plot. 

Complex and represents more information. It takes time to understand the plotting of data

```python
sns.violinplot(x = 'day' , y = 'total_bill' , hue = 'smoker' ,data = tips)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
sns.violinplot(x = 'day' , y = 'total_bill' , hue = 'sex', palette= 'husl', split= True,gridsize=200,data = tips)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
plt.grid()
```

```python
sns.violinplot(x = 'day' , y = 'total_bill' , hue = 'sex' , data = tips)
```

## Count Plot

A special case for the bar plot is when we want to show the number of observations in each category rather than computing a statistic for a second variable, such as mean, std etc. This is similar to a histogram over a categorical, rather than quantitative, variable. In seaborn, its easy to do so with the countplot() function.
This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences.

```python
sns.countplot(x= 'day'   , data = tips)
```

```python
tips.head()
```

```python
sns.countplot(hue = 'sex' , x = 'tip'  , data = tips)
```

```python
sns.countplot(hue = 'sex' , x = 'time'  , data = tips)
```

```python
sns.countplot(x = 'size'  , data = tips)
```
